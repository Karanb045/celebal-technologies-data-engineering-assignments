**Step 3**

```python
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
)

df['final_price'] = pd.to_numeric(df['final_price'])
```

```python
# Fill numerical column
df['discount'] = df['discount'].fillna(df['discount'].median())
# Fill text columns
text_cols = [
    'what_customers_said',
    'seller_name',
    'seller_information',
    'videos',
    'variations'
]
for col in text_cols:
    df[col] = df[col].fillna('Not Available')
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

**Step 4**

```python
df['price_difference'] = (df['initial_price'] - df['final_price'])
```

```python
df['popularity_score'] = (df['rating'] * np.log1p(df['ratings_count']))
```

```python
df.head()
```

```python
print("PRICE STATISTICS")
print("Initial Price:")
print(df['initial_price'].describe())

print("\nFinal Price:")
print(df['final_price'].describe())

print("\nRATING STATISTICS")
print(df['rating'].describe())

print("\nDISCOUNT STATISTICS")
print(df['discount'].describe())
```

**Step 5**

```python
#Univariate Analysis
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=20)
plt.title("Rating Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['final_price'], bins=30)
plt.title("Final Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['category'].value_counts().head(10).plot(kind='bar')
plt.title("Top Categories")
plt.show()
```

```python
#Bivariate Analysis
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='ratings_count',
    y='rating',
    data=df
)
plt.title("Ratings Count vs Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='initial_price',
    y='final_price',
    data=df
)

plt.title("Initial Price vs Final Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='discount',
    y='popularity_score',
    data=df
)

plt.title("Discount vs Popularity")
plt.show()
```

```python
#Category-level Analysis
category_rating = (
    df.groupby('category')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
category_rating.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title("Top Categories by Rating")
plt.show()
```

```python
category_price = (
    df.groupby('category')['final_price']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
category_price.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title("Most Expensive Categories")
plt.show()
```

```python
category_popularity = (
    df.groupby('category')['popularity_score']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
category_popularity.plot(
    kind='bar',
    figsize=(10,5)
)
plt.title("Most Popular Categories")
plt.show()
```

**Step 6**

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x=df['final_price']
)
plt.title("Price Outliers")
plt.show()
```

```python
numeric_cols = [
    'rating',
    'ratings_count',
    'initial_price',
    'discount',
    'final_price',
    'price_difference',
    'popularity_score'
]
plt.figure(figsize=(10,6))
sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix")
plt.show()
```

Key Findings


1. Most products are well-rated, with an average rating of 3.62 and a median rating of 4.1, indicating generally positive customer feedback.
2. Product prices vary widely, ranging from ₹199 to ₹17,995, showing a mix of budget and premium products.
3. Discounts are common, with an average discount of 53.5%.
4. No duplicate records were found, indicating good data quality and several columns contain many missing values.
5. Fashion categories dominate the catalog, with Tops, Dresses, and Shirts having the highest product counts.
6. Popularity is driven mainly by ratings and review counts, while price and discount have little impact on popularity.

Business Implications
1. Focus on maintaining high product ratings and encouraging customer reviews, as they strongly influence product popularity.
2. Continue investing in high-performing fashion categories, which make up a large portion of the catalog.
